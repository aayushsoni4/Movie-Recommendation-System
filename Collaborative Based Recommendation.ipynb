```python
movieTitle = pd.read_csv('dataset/tmdb_5000_movies.csv')
movieTitle = movieTitle[['id', 'title']]
```

```python
movieTitle
```

```python
title_id = dict(zip(movieTitle['title'], movieTitle['id']))
id_title = dict(zip(movieTitle['id'], movieTitle['title']))
```

```python
title_id['The Avengers']
```

```python
id_title[24428]
```

```python
the_avengers_corr = list(item_corr_matrix.loc[title_id['The Avengers']].items())
the_avengers_corr = sorted(the_avengers_corr, key=(lambda x: x[1]), reverse=True)
the_avengers_corr
```

```python
def recommendation(title):
    simi = list(item_corr_matrix.loc[int(title_id[title])].items())
    simi = sorted(simi, key=(lambda x: x[1]), reverse=True)
    movies = [id_title[int(float(i[0]))] for i in simi]
    return movies
```

```python
recommendation('The Avengers')
```

```python
recommendation('Star Wars')
```
